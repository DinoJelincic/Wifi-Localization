```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "../src/data/"
a507 = pd.read_csv(path + 'A507.csv')
hall = pd.read_csv(path + 'Hall.csv')
```

```python
def boxplot(csv):
    ap = list(csv.ap_id.unique())
    rssi = [csv.loc[csv.ap_id == i].rssi for i in ap]
    plt.figure(figsize=(14,8))
    plt.boxplot(rssi);
    plt.xticks(range(1,len(ap)+1),ap, rotation=60);
    plt.title(str(csv.room_id[0]))
    plt.show();
```

```python
def histograms(csv1, csv2):
    fig, ax = plt.subplots(1,2, figsize=(12,5))
    ax[0].hist(csv1.rssi)
    ax[0].set_title(str(csv1.room_id[0]))
    ax[1].hist(csv2.rssi)
    ax[1].set_title(str(csv2.room_id[0]))
    ax[0].grid()
    ax[1].grid()
    plt.tight_layout()
    plt.show();
```

```python
def grouped_histogram(csv1, csv2):
    fig, ax = plt.subplots(1,1, figsize=(8,6))
    ax.hist(csv1.rssi, color= 'blue', alpha=0.7, label=str(csv1.room_id[0]))
    ax.hist(csv2.rssi, color= 'red', alpha=0.7, label=str(csv2.room_id[0]))
    ax.set_title(f'{str(csv1.room_id[0])} and {str(csv2.room_id[0])}')
    ax.grid()
    ax.legend(loc='best')
    plt.show();
```

```python
for csv in [a507, hall]:
    print(f'{str(csv.room_id[0])} - There is {len(csv.ap_id.unique())} Access points.')
```

```python
histograms(a507,hall)
```

```python
grouped_histogram(a507,hall)
```
